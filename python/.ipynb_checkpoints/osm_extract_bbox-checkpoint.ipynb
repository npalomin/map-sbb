```python
import osmnx as ox
import pandas as pd
import geopandas as gpd
import requests

import os
from shapely.geometry import box
import matplotlib.pyplot as plt
```

```python
area_names = ['barcelona']
s_areas = ox.geocode_to_gdf(area_names)
```

```python
# total bound
sa_box = gpd.GeoSeries(box(*s_areas.total_bounds))
```

```python
ax = s_areas.plot()
sa_box.boundary.plot(color='r', ax=ax)
plt.show()
```

```python
area_names = ['new york']
s_areas = ox.geocode_to_gdf(area_names)
sa_box = gpd.GeoSeries(box(*s_areas.total_bounds))
ax = s_areas.plot()
sa_box.boundary.plot(color='r', ax=ax)
plt.show()
```
